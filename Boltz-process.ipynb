```python
import os, glob, json
import pandas as pd

# üìÇ Directorios
base_dir = "/content/drive/MyDrive/Biosensores/Semana_7/result-prueba2"
results_dir = os.path.join(base_dir, "results-boltz2")
rankings_dir = os.path.join(base_dir, "rankings")

# Crear carpeta rankings si no existe
os.makedirs(rankings_dir, exist_ok=True)

# Buscar todos los archivos de confianza dentro de results-boltz2
confidence_files = glob.glob(
    os.path.join(results_dir, "**/confidence_*_model_0.json"), recursive=True
)
print(f"üìÑ Se encontraron {len(confidence_files)} archivos de confianza")

records = []

for path in confidence_files:
    with open(path) as f:
        data = json.load(f)

    target = path.split("/")[-4]  # ejemplo: 'pdb10_1'
    record = {
        "target": target,
        "confidence_score": data.get("confidence_score"),
        "ptm": data.get("ptm"),
        "iptm": data.get("iptm"),
        "ligand_iptm": data.get("ligand_iptm"),
        "protein_iptm": data.get("protein_iptm"),
        "complex_plddt": data.get("complex_plddt"),
        "complex_iplddt": data.get("complex_iplddt"),
        "complex_pde": data.get("complex_pde"),
        "complex_ipde": data.get("complex_ipde"),
    }

    chains = data.get("chains_ptm", {})
    for k, v in chains.items():
        record[f"chain_{k}_ptm"] = v

    records.append(record)

# Crear dataframe y guardar CSV
df_conf = pd.DataFrame(records)
out_path = os.path.join(rankings_dir, "confidence_summary.csv")
df_conf.to_csv(out_path, index=False)

print(f"\n‚úÖ Resumen de confianza guardado en:\n{out_path}\n")
print(df_conf.head(10))
```

```python
import os, glob, json
import pandas as pd

# üìÇ Directorios
base_dir = "/content/drive/MyDrive/Biosensores/Semana_7/result-prueba2"
results_dir = os.path.join(base_dir, "results-boltz2")
rankings_dir = os.path.join(base_dir, "rankings")

# Crear carpeta rankings si no existe
os.makedirs(rankings_dir, exist_ok=True)

# Buscar todos los archivos de afinidad
affinity_files = glob.glob(
    os.path.join(results_dir, "**/affinity_*.json"), recursive=True
)
print(f"üìÑ Se encontraron {len(affinity_files)} archivos de afinidad")

records = []

for path in affinity_files:
    with open(path) as f:
        data = json.load(f)

    target = path.split("/")[-4]  # ejemplo: 'pdb10_1'
    record = {
        "target": target,
        "affinity_pred_value": data.get("affinity_pred_value"),
        "affinity_probability_binary": data.get("affinity_probability_binary"),
        "affinity_pred_value1": data.get("affinity_pred_value1"),
        "affinity_probability_binary1": data.get("affinity_probability_binary1"),
        "affinity_pred_value2": data.get("affinity_pred_value2"),
        "affinity_probability_binary2": data.get("affinity_probability_binary2"),
    }
    records.append(record)

# Crear DataFrame
df_aff = pd.DataFrame(records)

# Guardar CSV
out_path = os.path.join(rankings_dir, "affinity_summary.csv")
df_aff.to_csv(out_path, index=False)

print(f"\n‚úÖ Resumen de afinidad guardado en:\n{out_path}\n")
print(df_aff.head(10))
```

```python
import pandas as pd
import os

# üìÇ Directorio base
base_dir = "/content/drive/MyDrive/Biosensores/Semana_7/result-prueba2"
rankings_dir = os.path.join(base_dir, "rankings")

# Rutas de entrada
aff_path = os.path.join(rankings_dir, "affinity_summary.csv")
conf_path = os.path.join(rankings_dir, "confidence_summary.csv")

# Leer los CSV
df_aff = pd.read_csv(aff_path)
df_conf = pd.read_csv(conf_path)

print(f"‚úÖ Le√≠dos {len(df_aff)} registros de afinidad y {len(df_conf)} de confianza.")

# üîó Combinar por target
df_combined = pd.merge(df_aff, df_conf, on="target", how="inner")

# üßÆ Calcular puntaje combinado (ajusta los pesos si lo deseas)
aff_weight = 0.7
conf_weight = 0.3

# Escalar ambas m√©tricas para compararlas (evita que la afinidad domine)
df_combined["affinity_scaled"] = (
    df_combined["affinity_pred_value"] - df_combined["affinity_pred_value"].min()
) / (
    df_combined["affinity_pred_value"].max() - df_combined["affinity_pred_value"].min()
)

df_combined["confidence_scaled"] = (
    df_combined["confidence_score"] - df_combined["confidence_score"].min()
) / (
    df_combined["confidence_score"].max() - df_combined["confidence_score"].min()
)

# Puntaje ponderado (menor afinidad y mayor confianza = mejor)
df_combined["combined_score"] = (
    (1 - df_combined["affinity_scaled"]) * aff_weight
    + df_combined["confidence_scaled"] * conf_weight
)

# Orden descendente: mejor puntaje primero
df_ranked = df_combined.sort_values(by="combined_score", ascending=False)

# Guardar ranking
out_path = os.path.join(rankings_dir, "ranking_affinity_confidence_combined.csv")
df_ranked.to_csv(out_path, index=False)

print(f"\nüèÜ Ranking combinado guardado en:\n{out_path}\n")
print(df_ranked[["target", "affinity_pred_value", "confidence_score", "combined_score"]].head(10))
```

```python
import os, shutil, glob

# üìÇ Rutas base
project_dir = "/content/drive/MyDrive/Biosensores/Semana_7/result-prueba2"
results_dir = os.path.join(project_dir, "results-boltz2")
rankings_dir = os.path.join(project_dir, "rankings")
os.makedirs(rankings_dir, exist_ok=True)

# üßæ Tu lista (puede venir con o sin 'boltz_results_' al inicio)
names_raw = [
    "boltz_results_pdb19_1",
    "boltz_results_pdb17_2",
    "boltz_results_pdb27_1",
    "boltz_results_pdb19_3",
    "boltz_results_pdb19_5",
    "boltz_results_pdb29_5",
    "boltz_results_pdb27_5",
    "boltz_results_pdb19_4",
    "boltz_results_pdb19_2",
    "boltz_results_pdb4_3",
    "boltz_results_pdb16_3",
    "boltz_results_pdb25_5",
    "boltz_results_pdb4_5",
    "boltz_results_pdb28_5",
    "boltz_results_pdb12_5",
    "boltz_results_pdb28_4",
    "boltz_results_pdb1_2",
    "boltz_results_pdb28_3",
    "boltz_results_pdb30_3",
    "boltz_results_pdb6_1",
    "boltz_results_pdb28_2",
    "boltz_results_pdb31_2",
    "boltz_results_pdb27_3",
    "boltz_results_pdb27_4",
    "boltz_results_pdb2_1",
    "boltz_results_pdb32_4",
    "boltz_results_pdb27_2",
    "boltz_results_pdb32_5",
    "boltz_results_pdb17_5",
    "boltz_results_pdb13_5",
    "boltz_results_pdb17_3",
    "boltz_results_pdb29_3",
    "boltz_results_pdb4_1",
    "boltz_results_pdb5_2",
    "boltz_results_pdb6_3",
    "boltz_results_pdb16_4",
    "boltz_results_pdb30_4",
    "boltz_results_pdb5_1",
    "boltz_results_pdb16_2",
]

# üßπ Normalizar a targets tipo 'pdb10_1'
targets = [n.replace("boltz_results_", "") for n in names_raw]

copied, missing = [], []

for target in targets:
    # Ruta directa seg√∫n patr√≥n conocido
    direct_path = os.path.join(
        results_dir, target, f"boltz_results_{target}", "predictions", target, f"{target}_model_0.cif"
    )
    found_path = None
    if os.path.exists(direct_path):
        found_path = direct_path
    else:
        # B√∫squeda de respaldo (por si la carpeta intermedia tiene otro nombre o hay variaciones)
        pattern = os.path.join(results_dir, "**", "predictions", target, f"{target}_model_0.cif")
        matches = glob.glob(pattern, recursive=True)
        if matches:
            found_path = matches[0]

    if found_path:
        dst = os.path.join(rankings_dir, os.path.basename(found_path))
        shutil.copy2(found_path, dst)
        copied.append(dst)
    else:
        missing.append(target)

print(f"üèÜ Copiados {len(copied)} archivos .mmcif a: {rankings_dir}\n")
if missing:
    print("‚ö†Ô∏è No se encontraron .mmcif para:")
    for m in missing:
        print("  -", m)
else:
    print("‚úÖ Se copiaron todos correctamente.")
```
